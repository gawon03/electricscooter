```python
%cd /content/drive/MyDrive/Colab\ Notebooks/

from glob import glob

#img_list = glob('/content/drive/MyDrive/dataset/export/*/*.jpg')
img_list = glob('./es_project/images/*/*.jpg')

print(len(img_list))
```

```python
from sklearn.model_selection import train_test_split

train_img_list, val_img_list = train_test_split(img_list, test_size=0.2, random_state=2000)

print(len(train_img_list), len(val_img_list))
```

```python
with open('/content/drive/MyDrive/Colab Notebooks/es_project/train.txt', 'w') as f:
  f.write('\n'.join(train_img_list) + '\n')

with open('/content/drive/MyDrive/Colab Notebooks/es_project/val.txt', 'w') as f:
  f.write('\n'.join(val_img_list) + '\n')
```
